```python
durationMax = df.duration.max()
durationMin = df.duration.min()
pdayMax = df.pdays.max()
pdayMin = df.pdays.min()
```

```python
df.pdays = df.pdays.apply(lambda v:(v - pdayMin)/(pdayMax-pdayMin))
```
